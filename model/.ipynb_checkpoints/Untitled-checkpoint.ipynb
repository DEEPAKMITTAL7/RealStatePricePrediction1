```python
'''
I have decide to keep a range and store a avg of it
 and going to drop rest for now'''
```

```python
def avg(x):
    t = x.split('-')
    if len(t)==2:
        return (float(t[0])+float(t[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
avg('2100')
```

```python
avg('2100 - 2500 ')
```

```python
avg('3400sqft')
```

```python
data['total_sqft'] = data['total_sqft'].apply(avg)
data
```

```python
data.drop(['price_sqft','1br_area'],axis='columns',inplace=True)
data
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['area_type'] = le.fit_transform(data['area_type'])
data['location'] = le.fit_transform(data['location'])
```

```python
data
```

```python
x = data.drop(['price'],axis='columns')
x
```

```python
y = data[['price']]
y
```

```python
'''now lets change the name of size column as well'''
```

```python
data.rename(columns={'size':'bhk'},inplace=True)
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
r = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),x,y,cv=r).mean()
```

```python
cross_val_score(Lasso(alpha=50,max_iter=100,tol=0.1),x,y).mean()
```

```python
cross_val_score(Ridge(alpha=50,max_iter=100,tol=0.1),x,y).mean()
```

```python
'''L2 Regression giving us best result/

but all the regression models are close enough'''
```

```python
def prediction(a,b,c,d,e,f):
    return model.predict([[a,b,c,d,e,f]])
```

```python
prediction(3,65,2,1056,2,1)
```

```python
prediction(0,185,3,1440,2,3)
```
